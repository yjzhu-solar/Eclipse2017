```python
import matplotlib.pyplot as plt
import sunpy
import sunpy.map
from sunpy.visualization import colormaps
from astropy.visualization import ImageNormalize, AsinhStretch, LogStretch
```

```python
file_193 = "../../src/AIA/AIA20170821_172752_0193.fits"
file_304 = "../../src/AIA/AIA20170821_172753_0304.fits"
```

```python
aia_193_map = sunpy.map.Map(file_193)
```

```python
fig = plt.figure(figsize=(8,8),constrained_layout=True)
ax = fig.add_subplot(projection=aia_193_map)
norm_193 = ImageNormalize(data = aia_193_map.data, vmin=1,vmax=5e3,stretch=LogStretch())
ax.imshow(aia_193_map.data, norm=norm_193,cmap="sdoaia193")
plt.savefig(fname="../../figs/digitize/myAIA193.png",format="png",dpi=300,bbox_inches="tight")
```

```python
aia_304_map = sunpy.map.Map(file_304)
```

```python
fig = plt.figure(figsize=(8,8),constrained_layout=True)
ax = fig.add_subplot(projection=aia_304_map)
norm_304 = ImageNormalize(data = aia_304_map.data, vmin=1,vmax=5e2,stretch=LogStretch())
ax.imshow(aia_304_map.data, norm=norm_304,cmap="sdoaia304")
ax.scatter(0,0,transform=ax.get_transform("world"),marker="x")
plt.savefig(fname="../../figs/digitize/myAIA304.png",format="png",dpi=300,bbox_inches="tight")
```
