```python
import sunpy 
import sunpy.map
import matplotlib.pyplot as plt
import numpy as np
import sunkit_image.enhance as enhance
from astropy.visualization import (ImageNormalize, LinearStretch, SqrtStretch, LogStretch,
                                    AsinhStretch, MinMaxInterval, PercentileInterval)
```

```python
aia_171_map = sunpy.map.Map("../../src/AIA/AIA20170821_174645_0171.fits")
aia_211_map = sunpy.map.Map("../../src/AIA/AIA20170821_174645_0211.fits")
```

```python
aia_171_data_mgn = enhance.mgn(aia_171_map.data.astype(np.float64))
aia_171_map_mgn = sunpy.map.Map(aia_171_data_mgn, aia_171_map.meta)

aia_211_data_mgn = enhance.mgn(aia_211_map.data.astype(np.float64))
aia_211_map_mgn = sunpy.map.Map(aia_211_data_mgn, aia_211_map.meta)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection=aia_171_map_mgn)
aia_171_map_mgn.plot_settings["norm"] = ImageNormalize(vmin=0,vmax=1,stretch=SqrtStretch())
aia_171_map_mgn.plot(axes=ax)
plt.colorbar()
```
